```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
 #Ver los datos en películas
movies.head()

 # Ver todos los nombres de columna en películas y almacenarlos en forma de diccionario
movies.columns

 ## Ver datos en créditos
credits.head()

 # Ver todos los nombres de columna en créditos, almacenados en forma de diccionario, un total de 4 nombres de columna
credits.columns

 ##La columna de identificación en #movies es en realidad equivalente a la columna movie_id en créditos, que se puede usar como la clave principal para fusionar el marco de datos
full = pd.merge(movies, credits, left_on='id', right_on='movie_id', how='left')

 # Algunas columnas están fuera del alcance de este estudio, elimínelas
full.drop(['homepage','original_title','overview','spoken_languages',
           'status','tagline'],axis=1,inplace=True)

 #Ver información de datos, la cantidad de datos en cada campo.
full.info()
```

```python
movies.columns
```

```python
json_column = ['genres','keywords','production_companies',
               'production_countries','cast','crew']
```

```python
#Determinar la existencia de valores faltantes
full.isnull().any()
```

```python
full.loc[full['release_date'].isnull()==True]
```

```python
#Se busca la fecha faltantes de la pelicula , para insertarla
full['release_date'] = full['release_date'].fillna('2014-06-01')
```

```python
#Se evalua que falta en la columna runtime, para insertala
full.loc[full['runtime'].isnull()==True]
```

```python
full ['runtime'] = full ['runtime']. fillna (94, limit = 1) # limit = 1, limit para llenar solo un valor a la vez
full['runtime'] = full['runtime'].fillna(240, limit=1)
```

```python
#Cambiar la columna "release_date" a otra lectura
full['release_date'] = pd.to_datetime(full['release_date'],
    format='%Y-%m-%d', errors='coerce').dt.year
```

```python
#Convierta cada columna json a la lista del diccionario
for column in json_column:
    full[column]=full[column].map(json.loads)
```

```python
#Saque el valor correspondiente a la clave ‘nombre’ en el diccionario y genere una cadena separada por '|'
def getname(x):
    list = []
    for i in x:
        list.append(i['name'])
    return '|'.join(list)
```

```python
#Realizar funciones en los géneros, palabras clave, compañías de producción, columnas de países de producción.
for column in json_column[0:4]:
    full[column] = full[column].map(getname)
```

```python
#Defina una función para extraer 5 actores protagonistas:
def getcharacter(x):
    list = []
    for i in x:
        list.append(i['character'])
    return '|'.join(list[0:5])
```

```python
full['cast']=full['cast'].map(getcharacter)
```

```python
#Extraer el nomnbre del director
def getdirector(x):
    list=[]
    for i in x:
        if i['job']=='Director':
            list.append(i['name'])
    return "|".join(list)
```

```python
full['crew']=full['crew'].map(getdirector)
```

```python
rename_dict = {'release_date':'year','cast':'actor','crew':'director'}
full.rename(columns=rename_dict, inplace=True)
```

# Vamos a generar un nuevo conjunto, utilizando copias de los datos para poder tener mejores correciones

```python
genre_set = set () #Set conjunto vacío
for x in full['genres']:
         genre_set.update (x.split ('|')) #genres de datos están separados por '|'
```

```python
genre_df = pd.DataFrame () # Crear un marco de datos vacío
for genre in genre_set:
         # Si un valor contiene contenido específico, la codificación es 1, de lo contrario la codificación es 0
    genre_df[genre] = full['genres'].str.contains(genre).map(lambda x:1 if x else 0)
```

```python
genre_df['year']=full['year']
```

```python
#Agrupe genre_df por año y calcule la suma de cada grupo.
#Después de groupby, la columna del año se convierte automáticamente
#a df.index por el parámetro predeterminado as_index = True
```

```python
genre_by_year = genre_df.groupby('year').sum()
genresum_by_year = genre_by_year.sum().sort_values(ascending=False)
```

# **Rentabilidad de películas de diferentes estilos.**



```python
#PROFIT=GANANCIAS
full['profit'] = full['revenue']-full['budget']
```

```python
profit_df = pd.DataFrame () # Crear un marco de datos vacío
profit_df = pd.concat ([genre_df.iloc [:,: - 1], full ['profit']], axis = 1) #Merge
profit_df.head () # Ver nueva información del marco de datos
```

```python
#Una serie cuyo índice es cada género, y el valor es la suma de las ganancias calculadas por género
profit_by_genre = pd.Series(index=genre_set)
for genre in genre_set:
    grouped = profit_df.loc[:,[genre,'profit']].groupby(genre, as_index=False).sum()
    if len(grouped) > 1:
        profit_by_genre.loc[genre] = grouped.loc[1,'profit']
    else:
        profit_by_genre.loc[genre] = 0
print(profit_by_genre)
```

```python
#Crear una serie cuyo índice sea de cada género y el valor sea la suma del presupuesto calculado por género.
budget_df = pd.concat([genre_df.iloc[:,:-1],full['budget']],axis=1)
budget_df.head(2)
budget_by_genre = pd.Series(index=genre_set)
for genre in genre_set:
    grouped = budget_df.loc[:,[genre,'budget']].groupby(genre, as_index=False).sum()
    if len(grouped) > 1:
        budget_by_genre.loc[genre] = grouped.loc[1,'budget']
    else:
        budget_by_genre.loc[genre] = 0
print(budget_by_genre)
```

```python
profit_rate = pd.concat([profit_by_genre, budget_by_genre],axis=1)
profit_rate.columns = ['profit', 'budget'] #Cambiar nombre de columna
```

```python
#Columna de rendimiento
profit_rate['profit_rate'] = (profit_rate['profit']/profit_rate['budget'])*100
profit_rate.sort_values(by=['profit','profit_rate'], ascending=False, inplace=True)
print(profit_rate)
```

```python
x = list(range(len(profit_rate.index)))
```

```python
xl = list(profit_rate.index)
```

```python
#Visualice las ganancias (histograma) y el rendimiento (gráfico de líneas) de películas de diferentes estilos
fig = plt.figure(figsize=(18,13))
ax1 = fig.add_subplot(111)
plt.bar(x,profit_rate['profit'],label='Ganancias',alpha=0.7)
plt.xticks(x,xl,rotation=60,fontsize=12)
plt.yticks(fontsize=12)
ax1.set_title('Ganancias por genero', fontsize=20)
ax1.set_ylabel('Ganancias por filme',fontsize=18)
ax1.set_xlabel('Generos',fontsize=18)
ax1.set_ylim(0,1.2e11)
ax1.legend(loc=2,fontsize=15)


plt.bar(x, profit_rate['profit'], color=['red', 'green', 'blue','yellow','pink'], label='Ganancias', alpha=0.7)

```
