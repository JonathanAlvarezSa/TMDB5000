```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt


# Cargar los datos en un dataframe
df = movies

# Convertir las listas en cadenas de texto
df['production_companies'] = df['production_companies'].apply(lambda x: ','.join(map(str, x)))
df['genres'] = df['genres'].apply(lambda x: ','.join(map(str, x)))

# Seleccionar las variables categóricas que quieres convertir en numéricas
df_cat = df[['production_companies', 'genres']]

# Crear una instancia de OrdinalEncoder
encoder = OrdinalEncoder()

# Transformar las variables categóricas en numéricas
df_cat = pd.DataFrame(encoder.fit_transform(df_cat), columns=df_cat.columns)

# Unir el dataframe original con el dataframe de variables numéricas
df = pd.concat([df.drop(['production_companies', 'genres'], axis=1), df_cat], axis=1)

sns.boxplot(y='production_companies',x='genres',  data=df.head(100))
plt.gcf().set_size_inches(12, 8)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar los datos en un dataframe
df = movies

# Seleccionar los primeros 10 datos del dataframe
df_10 = df.head(10)

# Crea un gráfico de barras con un tamaño más grande
plt.figure(figsize=(24, 9))

# Crear el gráfico de barras utilizando seaborn
sns.barplot(x='movie_id', y='runtime', data=df.head(10))

# Agregar etiquetas de valor al gráfico de barras
def add_value_label(ax):
    for rect in ax.patches:
        x_value = rect.get_x() + rect.get_width() / 2
        y_value = rect.get_y() + rect.get_height()
        label = "{:.2f}".format(rect.get_height())
        ax.annotate(label, (x_value, y_value), xytext=(0, 5), textcoords="offset points", ha='center')

add_value_label(plt.gca())

plt.show()
```
